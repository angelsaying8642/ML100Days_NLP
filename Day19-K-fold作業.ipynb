## 作業目標：運用scikit-learn API 實現K-fold分割資料

---

### 讀入資料

```python
import pandas as pd
dataset = pd.read_csv(r'Social_Network_Ads.csv')
```

```python
dataset
```

### 取出訓練特徵與標註

```python
dataset[['User ID', 'Gender', 'Age', 'EstimatedSalary']].values
```

```python
X = dataset[['User ID', 'Gender', 'Age', 'EstimatedSalary']].values
Y = dataset['Purchased'].values
```

---

```python
import numpy as np
from sklearn.model_selection import KFold
```

### 將訓練資料按照順序切割成10等分

```python
# 預設為5
kf = KFold()
kf.get_n_splits(X)

print(kf)
```

```python

kf = KFold(10)
kf.get_n_splits(X)

print(kf)
```

### 將訓練資料隨機切割成10等分

```python
kf = KFold(10, shuffle=True)
kf.get_n_splits(X)

print(kf)
```

```python
kf = KFold("自行填入")
kf.get_n_splits(X)

print(kf)
```

---

### 取出 切割資料對應位置

```python
train_split = kf.split(X)
next(train_split)
```

### Or

```python
for train_index, test_index in kf.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
```

### 取出切割資料：trainset / testset 特徵(x_train/x_test)/標註(y_train/y_test)

```python
train_index, test_index = kf.split(X)
```

```python
X
```

```python
for train_index, test_index in kf.split(X):
    print("X_TRAIN:", X[train_index], "X_TEST:", X[test_index])
    print("Y_TRAIN:", Y[train_index], "Y_TEST:", Y[test_index])
```
